```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_csv('sales.csv')
df.head()
```

### 1. Check NULL values

```python
df.isnull().sum()
```

### 2. Removing duplicate rows with uncertain values

```python
data = []

for i in df.values:
    if i[2] != "Quantity Ordered":
        data.append(i)
```

```python
data = pd.DataFrame(data, columns=['order_id', 'product', 'quantity', 'price', 'date', 'address', 'month'])
```

```python
data.head()
```

### 3. Add total billing amount

```python
total = []

for i in data.values:
    qn = i[2]
    pr = i[3]
    total.append(round(int(qn)*float(pr), 2))
    
data['total'] = total
data.head()
         
```

### 4. Date/Time

```python
date_ = []
time_ = []

for i in data['date']:
    
    date_.append(i.split(' ')[0])
    time_.append(i.split(' ')[1])
    
data['date'] = date_
data['time'] = time_


data.head()
```

### 5. Hours/Minutes

```python
hour = []
for i in data['time']:
    hour.append(int(i.split(':')[0]))
    
data['hour'] = hour
data.head()
```

### 6. Adding Date, Month, August

```python
date = []
month = []
year = []

for i in data['date']:
    month.append(int(i.split('/')[0]))
    date.append(int((i.split('/')[1])))
    year.append(int(i.split('/')[2]))
    
data['date_'] = date
data['month_'] = month
data['year_'] = year

data.head()
```

### 7. Adding Day column

```python
datetime.date(1998, 10, 29).strftime("%a")
```

```python
day = []

for i in data.values:
    date = int(i[10])
    month = int(i[11])
    year = int(i[12])
    day.append(datetime.date(year, month, date).strftime('%a'))
    
data['day'] = day
data.head()
```

### 8. Address

```python
data['address']
```

#### 8.1 Pincode

```python
pincode = []
for i in data['address']:
    pincode.append(i.split(' ')[-1])
    
data['pincode'] = pincode
data.head()
```

#### 8.2 Validate pincode

```python
pincode = []

for i in data['pincode']:
    if (len(i) == 5):
        pincode.append(int(i))
        
data['pincode'] = pincode
data.head()
```

#### 8,3 Adding states column

```python
states = []
for i in data['address']:
    states.append(i.split(' ')[-2])

data['states'] = states
data.head()
```

#### 8.4  Adding city column

```python
city = []
for i in data['address']:
    city.append((i.split(' ')[-3]).split(',')[0])
    
data['city'] = city
data.head()
```

#### 8.5 Adding updated address

```python
address = []
for i in data['address']:
    address.append(i.split(',')[0].strip())

data['address'] = address
data.head()
```

### 9. Timezone

```python
timezone = []

for i in data['hour']:
    if i<=12:
        timezone.append('AM')
    else:
        timezone.append('PM')
```

```python
data['timezone'] = timezone
```

### 10. Saving the dataset

#### 9.1 Rearrange columns

```python
data['full_date'] = data['date']
del data['date']
```

```python
year = []
for i in data['year_']:
    year.append(int('20'+str(i)))
    
data['year'] = year
del data['year_']
data.head()
```

```python
df = pd.DataFrame()
```

```python
df['order_id']  =    data['order_id']
df['product']   =    data['product']
df['quantity']  =    data['quantity']
df['price']     =    data['price']
df['total']     =    data['total']
df['full_date'] =    data['full_date']   
df['hour']      =    data['hour']
df['date_']     =    data['date_']
df['month_']    =    data['month_']
df['year']      =    data['year']
df['day']       =    data['day']
df['address']   =    data['address']
df['city']      =    data['city']
df['states']    =    data['states']
df['pincode']   =    data['pincode']
df['timezone']  =    data ['timezone']
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.to_csv('data.csv', index=False)
```
