```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')
```

# 1. Price

```python
print('Average: ', int(df['price'].mean()))
print('Maximum: ', int(df['price'].max()))
print('Minimum: ', int(df['price'].min()))
```

```python
price = []

for i in df['price'].unique():
    price.append([i,len(df[df['price'] == i])])

df_pr = pd.DataFrame(price, columns=['price', 'frequency'])
df_pr.head()
```

```python
pr = list(df_pr.sort_values(by='price')['price'])
fr = list(df_pr.sort_values(by='price')['frequency'])
```

```python
pri = []
for i in pr:
    pri.append(str(i))
```

```python
fig, ax = plt.subplots(figsize=(13, 3))
ax.bar(pri, fr)
plt.xlabel('Price')
plt.ylabel("Purchase")
plt.title("Price Distribution")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.hist(df['price'], bins=5)
plt.show()
```
